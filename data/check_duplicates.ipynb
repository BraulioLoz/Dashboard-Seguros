```python
import pandas as pd

df = pd.read_csv("insurance_claims_clean.csv")

# Based on your get_or_create_insured logic in add_data.py
# Your code uses: age, sex, education_level, occupation, zip_code
insured_duplicates = df.groupby([
    'age',                    # Matching field
    'insured_sex',           # Matching field (not hobbies/relationships!)
    'insured_education_level', # Matching field
    'insured_occupation',     # Matching field
    'insured_zip'            # Matching field (NOT months_as_customer!)
]).size()

print("üîç DUPLICATE ANALYSIS")
print("=" * 50)

print(f"\nüìä INSURED:")
print(f"Total rows: {len(df)}")
print(f"Unique insureds: {len(insured_duplicates)}")
print(f"Insureds appearing more than once: {(insured_duplicates > 1).sum()}")

if (insured_duplicates > 1).sum() > 0:
    print("\nDuplicated insureds:")
    print(insured_duplicates[insured_duplicates > 1].sort_values(ascending=False))

# POLICY duplicates
policy_duplicates = df.groupby('policy_number').size()
print(f"\nüìä POLICY:")
print(f"Unique policies: {len(policy_duplicates)}")
print(f"Policies appearing more than once: {(policy_duplicates > 1).sum()}")

if (policy_duplicates > 1).sum() > 0:
    print("Duplicated policies:")
    print(policy_duplicates[policy_duplicates > 1].sort_values(ascending=False))

# VEHICLE duplicates
vehicle_duplicates = df.groupby(['auto_make', 'auto_model', 'auto_year']).size()
print(f"\nüìä VEHICLE:")
print(f"Unique vehicles: {len(vehicle_duplicates)}")
print(f"Vehicles appearing more than once: {(vehicle_duplicates > 1).sum()}")

if (vehicle_duplicates > 1).sum() > 0:
    print("Vehicles appearing more than once:")
    print(vehicle_duplicates[vehicle_duplicates > 1].sort_values(ascending=False))

print("\n" + "=" * 50)
print("SUMMARY:")
print(f"Cases: 1000 (each row is unique)")
print(f"Incidents: 1000 (one per case)")
print(f"Claims: 1000 (one per case)")
```
